# ðŸ§© SQL Commands

- SQL commands are instructions used to communicate with a database.
- They help you create, retrieve, update, and manage data in a relational database.



##  Types of SQL Commands

| Category                               | Description                            | Example Commands                                |
| -------------------------------------- | -------------------------------------- | ----------------------------------------------- |
| **DDL (Data Definition Language)**     | Defines or modifies database structure | `CREATE`, `ALTER`, `DROP`, `TRUNCATE`, `RENAME` |
| **DML (Data Manipulation Language)**   | Used to manipulate data in tables      | `INSERT`, `UPDATE`, `DELETE`                    |
| **DQL (Data Query Language)**          | Used to retrieve data                  | `SELECT`                                        |
| **DCL (Data Control Language)**        | Controls user access                   | `GRANT`, `REVOKE`                               |
| **TCL (Transaction Control Language)** | Manages database transactions          | `COMMIT`, `ROLLBACK`, `SAVEPOINT`               ||



# 1. **ðŸ“˜ What is DDL?**

- DDL (Data Definition Language) is used to define, modify, and manage the structure of database objects â€” like tables, schemas, and indexes.

-  DDL commands affect the structure of the database, not the data inside it.

| ------------------- | ------------------------------------------------------------------ |

| `SHOW`              | Displays information about databases, tables, columns, etc.        |
| `DESCRIBE` / `DESC` | Displays structure of a table                                      |

# 2. ðŸ“˜ What is DML?

- DML (Data Manipulation Language) is used to insert, update, delete, and retrieve data stored in database tables.
- These commands work inside tables â€” they manipulate data, not structure.
- DML commands can be rolled back using transactions (ROLLBACK, COMMIT).
- They affect rows (records) inside a table.

## SQL Keywords for Data Manipulation (DML)

| Keyword  | Description                             |
| -------- | --------------------------------------- |
| `INSERT` | Adds new rows into a table              |
| `UPDATE` | Modifies existing rows                  |
| `DELETE` | Removes rows from a table               |
| `VALUES` | Specifies values for `INSERT`           |
| `SET`    | Used with `UPDATE` to assign new values |


# ðŸ“˜ What is DQL?

- DQL (Data Query Language) is used to retrieve data from a database.
- It helps users query, filter, sort, and analyze information stored in tables.

- DQL mainly revolves around one powerful command: SELECT
- It does not modify data, only reads it.
- It can work with one or more tables at once.

## SQL Keywords for Data Query (DQL)

| Keyword    | Description                           |
| ---------- | ------------------------------------- |
| `SELECT`   | Fetches data from one or more tables  |
| `FROM`     | Specifies table name(s)               |
| `WHERE`    | Filters records based on condition    |
| `ORDER BY` | Sorts the result set (ASC or DESC)    |
| `GROUP BY` | Groups rows based on column values    |
| `HAVING`   | Filters groups (used with GROUP BY)   |
| `DISTINCT` | Removes duplicate rows                |
| `LIMIT`    | Restricts number of rows returned     |
| `TOP`      | Limits number of rows (in SQL Server) |
| `AS`       | Renames a column or table (alias)     |


# ðŸ“˜ What is DCL ?

- DCL stands for Data Control Language
- DCL commands in SQL are used to control access or permissions to the database objects like tables, views, and schemas.
- They help in managing user privileges and security within the database.

## SQL Keywords for Data Control (DCL)

| Keyword  | Description                      |
| -------- | -------------------------------- |
| `GRANT`  | Gives access rights to users     |
| `REVOKE` | Removes access rights from users |


# ðŸ“˜ What is TCL ?
 - TCL stands for Transaction Control Language

- TCL commands in SQL are used to manage transactions in a database.
- A transaction is a group of one or more SQL operations (like INSERT, UPDATE, DELETE) that are executed as a single unit of work.

- TCL ensures data integrity and allows you to commit or rollback changes made during a transaction

##  SQL Keywords for Transaction Control (TCL)

| Keyword       | Description                                |
| ------------- | ------------------------------------------ |
| `COMMIT`      | Saves all changes permanently              |
| `ROLLBACK`    | Undo changes since last COMMIT             |
| `SAVEPOINT`   | Creates a save point to rollback partially |
| `TRANSACTION` | Starts a new transaction                   |
| `BEGIN`       | Marks beginning of a transaction           |
| `END`         | Marks end of a transaction                 |

# ðŸ“˜ What is SQL Aggregate Functions ?

- Aggregate functions perform a calculation on a set of values and return a single value.
- They are often used with the GROUP BY clause to summarize data.


## SQL Aggregate Functions (used with SELECT)

| Keyword   | Description                   |
| --------- | ----------------------------- |
| `COUNT()` | Returns number of rows        |
| `SUM()`   | Adds up numeric column values |
| `AVG()`   | Returns average               |
| `MIN()`   | Returns smallest value        |
| `MAX()`   | Returns largest value         |

# ðŸ“˜ What is Sql Join ?

- A JOIN is used to combine rows from two or more tables based on a related column between them.

- Joins are mainly used in SELECT queries to fetch data from multiple tables.

- Commonly, the foreign key in one table matches the primary key in another table.


## SQL Keywords for Joins (Combining Tables)

| Keyword      | Description                                                |
| ------------ | ---------------------------------------------------------- |
| `JOIN`       | Combines rows from two or more tables                      |
| `INNER JOIN` | Returns records with matching values in both tables        |
| `LEFT JOIN`  | Returns all records from left table + matched from right   |
| `RIGHT JOIN` | Returns all records from right table + matched from left   |
| `FULL JOIN`  | Returns all records where there is a match in either table |
| `ON`         | Specifies join condition                                   |
| `UNION`      | Combines results of two SELECT queries (no duplicates)     |
| `UNION ALL`  | Combines results including duplicates                      |
